# Plot the data

```python
plt.figure(figsize=(10, 5))
plt.plot(df['close'])
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.title('AMZN Close Price')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['volume'])
plt.xlabel('Days')
plt.ylabel('Volume')
plt.title('AMZN Volume')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['open'])
plt.xlabel('Days')
plt.ylabel('Open Price')
plt.title('AMZN Open Price')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['high'])
plt.xlabel('Days')
plt.ylabel('High Price')
plt.title('AMZN High Price')
plt.show()
```
