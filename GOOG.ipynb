# plot the data set. GOOG_data.csv

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"], df["close"])
plt.xlabel("Date")
plt.ylabel("Close")
plt.title("Google Stock Price")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"], df["volume"])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Google Stock Volume")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"], df["open"])
plt.xlabel("Date")
plt.ylabel("Open")
plt.title("Google Stock Open")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"], df["high"])
plt.xlabel("Date")
plt.ylabel("High")
plt.title("Google Stock High")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"], df["low"])
plt.xlabel("Date")
plt.ylabel("Low")
plt.title("Google Stock Low")
plt.show()
```
